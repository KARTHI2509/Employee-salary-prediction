```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

# ===== 1. Load and preprocess data =====
data = pd.read_excel(r"C:\Users\karth\Downloads\employee\Employees.xlsx")

# Select features and target
x = data[["Years", "Job Rate", "Gender", "Department", "Country", "Center",
          "Sick Leaves", "Unpaid Leaves", "Overtime Hours","Monthly Salary"]].copy()
y = data["Annual Salary"]

# Encode categorical features safely
label_columns = ["Gender", "Department", "Country", "Center"]
label_encoders = {}

for col in label_columns:
    le = LabelEncoder()
    x.loc[:, col] = le.fit_transform(x[col])  # FIXED: avoid SettingWithCopyWarning
    label_encoders[col] = le  # Save for future decoding

# ===== 2. Train-test split =====
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# ===== 3. Define models =====
models = {
    "RandomForest": RandomForestRegressor(n_estimators=100, random_state=42),
    "DecisionTree": DecisionTreeRegressor(random_state=42),
    "KNN": KNeighborsRegressor(n_neighbors=5),
    "LinearRegression": LinearRegression()
}

# ===== 4. Train, evaluate and compare =====
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    r2 = r2_score(y_test, preds)
    mae = mean_absolute_error(y_test, preds)
    results[name] = {"model": model, "r2": r2, "mae": mae}
    print(f"{name} — R²: {r2:.3f}, MAE: {mae:.2f}")

# ===== 5. Save the best model =====
best_name = max(results, key=lambda x: results[x]["r2"])
best_model = results[best_name]["model"]

print(f"\n✅ Best Model: {best_name} (R² = {results[best_name]['r2']:.3f})")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print("💾 Saved best model as best_model.pkl")

# Optional: Save encoders
joblib.dump(label_encoders, "label_encoders.pkl")
print("💾 Saved label encoders as label_encoders.pkl")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
model_names = list(results.keys())
r2_scores = [results[m]["r2"] for m in model_names]

# Dynamic y-axis: Zoom in around your values
min_r2 = min(r2_scores)
max_r2 = max(r2_scores)
margin = 0.002  # small zoom margin
ymin = max(0, min_r2 - margin)
ymax = min(1, max_r2 + margin)

# Plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=model_names, y=r2_scores, palette="crest")

# Add value labels on bars
for i, v in enumerate(r2_scores):
    ax.text(i, v + 0.0005, f"{v:.4f}", ha='center', fontsize=11, fontweight='bold')

plt.title("📊 R² Score Comparison — Regression Models", fontsize=14)
plt.ylabel("R² Score")
plt.xlabel("Model")
plt.ylim(ymin, ymax)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
import pandas as pd
import joblib

# === Load model and label encoders ===
model = joblib.load("best_model.pkl")
encoders = joblib.load("label_encoders.pkl")

# === Page config ===
st.set_page_config(page_title="💼 Salary Prediction", page_icon="💰", layout="centered")

# === Sidebar Navigation ===
tab = st.sidebar.radio("🔘 Navigate", ["📊 Single Prediction", "📂 CSV Upload"])

# === Title ===
st.title("💼 Employee Salary Prediction")
st.markdown("Predict salary based on employee attributes or upload a CSV file.")
st.divider()

# === Tab 1: Single Prediction Form ===
if tab == "📊 Single Prediction":
    if "reset" not in st.session_state:
        st.session_state.reset = False

    def reset_form():
        st.session_state.reset = True

    with st.form("prediction_form"):
        st.subheader("📋 Employee Information")

        gender_options = encoders["Gender"].classes_
        department_options = encoders["Department"].classes_
        country_options = encoders["Country"].classes_
        center_options = encoders["Center"].classes_

        years = st.number_input("Years of Experience", value=1, step=1, min_value=0, key="years")
        jobrate = st.slider("Job Performance Rating", 0.0, 5.0, step=0.5, key="jobrate")

        col1, col2 = st.columns(2)
        with col1:
            gender = st.selectbox("Gender", gender_options, key="gender")
            department = st.selectbox("Department", department_options, key="department")
        with col2:
            country = st.selectbox("Country", country_options, key="country")
            center = st.selectbox("Center", center_options, key="center")

        sick_leaves = st.number_input("Sick Leaves Taken", min_value=0, value=0, step=1, key="sick_leaves")
        unpaid_leaves = st.number_input("Unpaid Leaves Taken", min_value=0, value=0, step=1, key="unpaid_leaves")
        overtime_hours = st.number_input("Overtime Hours", min_value=0, value=0, step=1, key="overtime_hours")
        monthly_salary = st.number_input("Monthly Salary", min_value=0, value=0, step=1000, key="monthly_salary")

        col_predict, col_reset = st.columns(2)
        submit = col_predict.form_submit_button("🎯 Predict Salary")
        reset = col_reset.form_submit_button("🔄 Reset", on_click=reset_form)

        if reset:
            st.rerun()

        if submit:
            gender_encoded = encoders["Gender"].transform([gender])[0]
            department_encoded = encoders["Department"].transform([department])[0]
            country_encoded = encoders["Country"].transform([country])[0]
            center_encoded = encoders["Center"].transform([center])[0]

            input_data = {
                "Years": years,
                "Job Rate": jobrate,
                "Gender": gender,
                "Department": department,
                "Country": country,
                "Center": center,
                "Sick Leaves": sick_leaves,
                "Unpaid Leaves": unpaid_leaves,
                "Overtime Hours": overtime_hours,
                "Monthly Salary": monthly_salary
            }

            st.info("📄 Entered Information")
            st.dataframe(pd.DataFrame([input_data]))

            features = np.array([[years, jobrate, gender_encoded, department_encoded, country_encoded,
                                  center_encoded, sick_leaves, unpaid_leaves, overtime_hours, monthly_salary]])

            prediction = model.predict(features)[0]
            st.success(f"🌸 Estimated Annual Salary: ₹{prediction:,.2f}")
            st.balloons()


# === Tab 2: CSV Upload ===
elif tab == "📂 CSV Upload":
    st.subheader("📂 Upload Employee CSV File")
    st.markdown("Make sure the CSV matches the format used for model training.")

    uploaded_file = st.file_uploader("Upload CSV File", type=["csv", "xlsx"])

    if uploaded_file:
        try:
            df = pd.read_csv(uploaded_file) if uploaded_file.name.endswith(".csv") else pd.read_excel(uploaded_file)

            st.info("📄 Uploaded File Preview")
            st.dataframe(df.head())

            for col in ["Gender", "Department", "Country", "Center"]:
                if col in df.columns:
                    df[col] = encoders[col].transform(df[col])

            expected_cols = ["Years", "Job Rate", "Gender", "Department", "Country",
                             "Center", "Sick Leaves", "Unpaid Leaves", "Overtime Hours", "Monthly Salary"]
            df = df[expected_cols]

            predictions = model.predict(df)
            df["Predicted Annual Salary"] = predictions
            st.success("✅ Batch predictions completed!")
            st.dataframe(df)

            csv = df.to_csv(index=False).encode('utf-8')
            st.download_button("📥 Download Results", csv, file_name="salary_predictions.csv", mime="text/csv")
        except Exception as e:
            st.error(f"❌ Error: {e}")
```
